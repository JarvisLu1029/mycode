```python
import pymysql

def read_data(sql_query, connection):
    with connection.cursor() as cursor:
        cursor.execute(sql_query)

        while True:
            records = cursor.fetchmany(size=100)  # 這裡可以調整每次獲取的記錄數量
            if not records:
                break
            for record in records:
                yield record

# 使用示例
connection = pymysql.connect(host='your_host', user='your_user', password='your_password', db='your_db')

for row in read_data("SELECT * FROM your_table", connection):
    print(row)

connection.close()
```

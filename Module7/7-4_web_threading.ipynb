```python
from web_crawling import *
import time, queue

def get_commodity(commodity):
    q = queue.Queue()
    start_time = time.time()
    momo_dict = momo_data(q, commodity)
    carrefour_dict = carrefour_data(q, commodity)
    pchome_dict = pchome_data(q, commodity)
    pxmart_dict = pxmart_data(q, commodity)
    end_time = time.time()

    execution_time = end_time - start_time
    print(f'共執行 {execution_time} 秒')
    print(momo_dict)
    print(carrefour_dict)
    print(pchome_dict)
    print(pxmart_dict)

get_commodity('衛生紙')
```

```python
def queue_consumer(web_q, result_q):
    total_dict = {}

    n = 0
    while True:
        data = web_q.get()
        if data == None:
            break

        for i in data:
            total_dict[f'{n}'] = {
                'commodity_name': data[i]['commodity_name'],
                'commodity_value': data[i]['commodity_value'],
                'commodity_link': data[i]['commodity_link'],
                'commodity_image': data[i]['commodity_image'],
                'store': data[i]['store']
            }
            n += 1
    
    result_q.put(total_dict)
    return total_dict
```

```python
from web_crawling import *
import threading, queue
from pprint import pprint

def multithread(commodity):
    tasks = [pxmart_data, momo_data, pchome_data, carrefour_data]
    q = queue.Queue()
    result_q = queue.Queue()
    all_thread = []

    for task in tasks:
        thread = threading.Thread(target=task, args=(q, commodity))
        thread.start()
        all_thread.append(thread)

    thread = threading.Thread(target=queue_consumer, args=(q, result_q))
    thread.start()

    for t in all_thread:
        t.join()
    
    q.put(None)
    thread.join()

    result = result_q.get()

    return result

start_time = time.time()
multithread('衛生紙')
end_time = time.time()

execution_time = end_time - start_time

print(f'共執行 {execution_time} 秒')
```

```python
from web_crawling import *
import time, queue

def get_commodity(commodity):
    q = queue.Queue()
    start_time = time.time()
    momo_dict = momo_data(q, commodity)
    carrefour_dict = carrefour_data(q, commodity)
    pchome_dict = pchome_data(q, commodity)
    pxmart_dict = pxmart_data(q, commodity)
    end_time = time.time()

    execution_time = end_time - start_time
    print(f'共執行 {execution_time} 秒')
    print(momo_dict)
    print(carrefour_dict)
    print(pchome_dict)
    print(pxmart_dict)

get_commodity('衛生紙')
```

```python
from web_crawling import *
import threading, queue
from pprint import pprint

def multithread(commodity):
    tasks = [pxmart_data, momo_data, pchome_data, carrefour_data]
    q = queue.Queue()
    all_thread = []

    for task in tasks:
        thread = threading.Thread(target=task, args=(q, commodity))
        thread.start()
        all_thread.append(thread)

    for t in all_thread:
        t.join()

    total_dict = {}
    n = 0
    while not q.empty():
        results = q.get()

        for i in results:
            total_dict[f'{n}'] = {
                'commodity_name': results[i]['commodity_name'],
                'commodity_value': results[i]['commodity_value'],
                'commodity_link': results[i]['commodity_link'],
                'commodity_image': results[i]['commodity_image']
            }
            n += 1
    pprint(total_dict)
    # pprint(results)

start_time = time.time()
multithread('衛生紙')
end_time = time.time()

execution_time = end_time - start_time

print(f'共執行 {execution_time} 秒')
```
